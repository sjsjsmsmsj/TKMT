Sau khi đọc dữ liệu từ file CSV vào DataFrame df, hãy thực hiện các yêu cầu sau:

- (2 điểm) Hiển thị 5 dòng đầu tiên, 5 dòng ngẫu nhiên, kích thước DataFrame, số dòng và số cột.
- (2 điểm) Thống kê số lượng xe theo từng hãng (HangXe) và vẽ biểu đồ cột minh họa.
- (2 điểm) Tính giá bán trung bình (GiaBan) của xe theo từng loại (LoaiXe).
- (2 điểm) Lọc ra những xe có dung tích trên 200cc và hiển thị.
- (2 điểm) Vẽ biểu đồ pie thể hiện tỷ lệ xe được bán theo từng loại (LoaiXe).


```python
import pandas as pd 
df = pd.read_csv('motorbike_store.csv')
df
```

```python
# 5 dòng đầu tiên
print("5 dòng đầu tiên: ", df.head())

# 5 dòng ngẫu nhiên 
print("5 dòng ngẫu nhiên: ", df.sample(5))

# kích thước 
print("Kích thước của dữ liệu:", df.shape)

# số dòng 
print("Số dòng: ", df.shape[0])

# số cột 
print("Số cột: ", df.shape[1])
```

```python
# thống kê số lượng hãng xe của từng hãng
hang_xe = df.groupby('HangXe')['HangXe'].count()
print(hang_xe)
# vẽ biểu đồ pie
import matplotlib.pyplot as plt 
plt.pie(hang_xe, labels=['Honda', 'Kawasaki', 'Suzuki', 'Yamaha'])
plt.legend()
plt.title("Biểu đồ thể hiện số lượng của từng loại xe")

plt.show()
```

Nhận xét:
- có ba hãng xe có số lượng bằng nhau 
- có 1 hãng xe đó là Kawasaki là chỉ chiếm bằng 1/3 với các hãng còn lại

```python
gia_ban = df.groupby('HangXe')['GiaBan'].mean()
print(gia_ban)
```

```python
dung_tich = df[df['DungTich'] > 200]['HangXe']
print(dung_tich)
```

```python
ban_tung_loai_xe = df.groupby('HangXe')['SoLuongBan'].sum()
print(ban_tung_loai_xe)
```

```python
import matplotlib.pyplot as plt 
plt.pie(ban_tung_loai_xe, labels=['Honda', 'Kawasaki', 'Suzuki', 'Yamaha'])
plt.legend()
plt.title("Biểu đồ thể hiện số lượng bán được của mỗi hãng xe")
plt.show()
```

Nhận xét: 
- Số lượng bán được của hãng xe Yamaha là nhiều nhất 
- Số lượng bán được của hãng xe Kawasaki là thấp nhất
