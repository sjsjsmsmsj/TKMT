- (2 điểm) Đọc dữ liệu vào DataFrame và hiển thị 5 dòng đầu, 5 dòng ngẫu nhiên, kích thước DataFrame.
- (2 điểm) Thống kê số lượng điện thoại bán ra theo từng hãng (HangDT) và vẽ biểu đồ cột.
- (2 điểm) Tính giá bán trung bình (GiaBan) theo từng loại (LoaiDT).
- (2 điểm) Lọc ra những điện thoại có dung lượng bộ nhớ trong (BoNhoTrong) lớn hơn 128GB và hiển thị.
- (2 điểm) Vẽ biểu đồ pie thể hiện tỷ lệ điện thoại bán ra theo từng loại (LoaiDT).


```python
import pandas as pd
df = pd.read_csv("smartphone_store.csv")

print("5 dòng đầu tiên:", df.head())
print("5 dòng ngẫu nhiên:", df.sample(5))
print("Kích thước:", df.shape)
print("Số dòng: ", df.shape[0])
print("Số cột:", df.shape[1])
```

```python
so_luong = df.groupby("HangDT")['HangDT'].count()
print(so_luong)
import matplotlib.pyplot as plt 
plt.bar(so_luong.index, so_luong.values)
plt.show()
```

```python
trung_binh = df.groupby('HangDT')['GiaBan'].mean()
print("Trung bình giá bán của mỗi hãng điện thoại:\n", trung_binh)
```

```python
bo_nho_trong = df[df['BoNhoTrong'] > 128]
print( bo_nho_trong)
```

```python
thong_ke = df.groupby('LoaiDT')['SoLuongBan'].sum()
print(thong_ke)
```

```python
plt.pie(thong_ke, labels=['Flagship', 'Giá rẻ', 'Tầm trung'])
plt.title("Biểu đồ thể hiện tỷ lệ bán ra theo từng loại điện thoại")
plt.legend()
plt.show()
```
