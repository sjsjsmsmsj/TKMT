Phần 1: Phân phối nhị thức (Binomial Distribution)
Một nhà sản xuất chip máy tính biết rằng xác suất để một con chip bị lỗi là 0.02. Nhà máy sản xuất 1000 con chip mỗi ngày.

(1 điểm) Tạo một biến ngẫu nhiên theo phân phối nhị thức với n = 1000, p = 0.02 và sinh 10,000 mẫu dữ liệu.

(1 điểm) Tính toán kỳ vọng và phương sai của phân phối này.

(1.5 điểm) Vẽ biểu đồ histogram của dữ liệu đã sinh.

(0.5 điểm) Tìm xác suất để có nhiều hơn 25 con chip lỗi trong một ngày.

```python
# đầu tiên cần import mấy cái thư viện 
from scipy.stats import binom, poisson, norm 
import matplotlib.pyplot as plt
```

```python
# bước 2 lật cái tờ giấy ra và nhìn vào phân phối nhị thức 
n = 1000 
p = 0.02
ngau_nhien = binom.rvs(n, p, size=10000)
```

```python
# lật cái tờ giấy ra tìm kỳ vọng và phương sai 
mu = binom.mean(n, p)
var = binom.var(n, p)
print(mu)
print(var)
```

```python
plt.hist(ngau_nhien)
plt.title("Biểu đồ histogram cho tập mẫu ngẫu nhiên")
plt.show()
```

```python
# P(x > 25) = 1 - P(x <= 25)
p_25 = binom.cdf(25, n, p)
p = 1 - p_25
print(p)
```

Phần 2: Phân phối Poisson (Poisson Distribution)
Một trạm xe buýt ghi nhận rằng trung bình có 4 xe buýt đến mỗi giờ.

(1 điểm) Mô phỏng một biến ngẫu nhiên có phân phối Poisson với trung bình λ = 4 và sinh 5000 giá trị.

(1 điểm) Tính toán kỳ vọng và phương sai của phân phối này.

(1.5 điểm) Vẽ biểu đồ histogram thể hiện tần suất số lượng xe buýt đến trong một giờ.

(0.5 điểm) Xác suất để trong một giờ có đúng 6 xe buýt đến là bao nhiêu?



```python
landa = 4
# bước 2 là nhìn vào poisson  trong tờ giấy
ngau_nhien = poisson.rvs(landa, size=5000)
```

```python
# nhìn vào tờ giấy thì ba cái này = nhau
m = landa 
phuong_sai = landa 
print(m)
print(phuong_sai)
```

```python
plt.hist(ngau_nhien)
plt.show()
```

```python
# p (x = 6)
p_6 = poisson.pmf(6, landa)
print(p_6)
```

Phần 3: Phân phối chuẩn (Normal Distribution)
Chiều cao của sinh viên trong một trường đại học được giả định là tuân theo phân phối chuẩn với trung bình 170 cm và độ lệch chuẩn 8 cm.

(1 điểm) Mô phỏng 10,000 giá trị của chiều cao sinh viên theo phân phối chuẩn.

(1 điểm) Tính trung bình, phương sai và độ lệch chuẩn của dữ liệu đã sinh.

(1.5 điểm) Vẽ biểu đồ histogram của chiều cao sinh viên và chồng lên đó đường cong mật độ xác suất (PDF).

(0.5 điểm) Tìm xác suất để một sinh viên có chiều cao lớn hơn 180 cm.



```python
trung_binh = 170 
do_lech_chuan = 8 
# nhìn vào tờ giấy phần phân phối chuẩn
ngau_nhien = norm.rvs(size=10000, loc=trung_binh, scale=do_lech_chuan)
```

```python
trung_binh_1 = ngau_nhien.mean() 
phuong_sai_1 = ngau_nhien.var()
do_lech_chuan_1 = ngau_nhien.std()
print(trung_binh_1)
print(phuong_sai_1)
print(do_lech_chuan_1)
```

```python
import numpy as np
# bước đầu tiên vẽ biểu đồ hist  
plt.hist(ngau_nhien, density=True) # vì là hàm mật độ nhìn vào tờ giấy ta có thể thấy density=True
x = np.linspace(130, 200, 100)
pdf_v = norm.pdf(x, loc=trung_binh, scale=do_lech_chuan)
plt.plot(x, pdf_v)
plt.show()
# trong tờ giấy thiếu plot 
# linspace ý nghĩa là (giá trị đi từ đoạn nào đến đoạn nào) chia làm 100 khoảng
# tờ giấy có chỉ cách vẽ :))) chỗ hàm mật độ
```

```python
# P(x > 180) = 1 - P(x <= 180)
p_180 = norm.cdf(180, loc=trung_binh, scale=do_lech_chuan)
p = 1 - p_180
print(p)
```
