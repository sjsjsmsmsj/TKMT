```python
bbts = df['Year'].value_counts().reset_index()
print(bbts)
```

```python
sv_tap_the_duc_3 = (df['Exercise'] < 3).sum()
tong_the_sv = len(df)
print(sv_tap_the_duc_3 / tong_the_sv)
```

```python
nam = df[df['Gender'] == 'Nam']
nam_ttd = (nam['Exercise'] < 3).sum()
ty_le = nam_ttd / len(nam)
print("Tỷ lệ nam tập thể dục ít hơn 3 lần : ", ty_le)
```

```python
nu = df[df['Gender'] == 'Nữ']
nu_ttd = (nu['Exercise'] < 3).sum()
ty_le = nu_ttd / len(nu)
print("Tỷ lệ nam tập thể dục ít hơn 3 lần : ", ty_le)
```

Nhận xét: 
- Cho thấy tỷ lệ nam cao hơn
