Nhận xét: 
- Cho thấy tỷ lệ nữ của dữ liệu gốc có xác xuất đúng hơn 
khi có càng nhiều dữ liệu

```python
trung_binh = []
for i in range(100):
    mau = df.sample(50)
    nu = mau[mau['Gender'] == 'Nữ']
    hut_thuoc_nu = (nu['Smoking'] == 1).sum()
    tong_the_nu = len(nu)
    ty_le =hut_thuoc_nu / tong_the_nu
    trung_binh.append(ty_le)
plt.hist(trung_binh)
plt.show()
```

```python
trung_binh = []
for i in range(100):
    mau = df.sample(150)
    nu = mau[mau['Gender'] == 'Nữ']
    hut_thuoc_nu = (nu['Smoking'] == 1).sum()
    tong_the_nu = len(nu)
    ty_le =hut_thuoc_nu / tong_the_nu
    trung_binh.append(ty_le)
plt.hist(trung_binh)
plt.show()
```

Nhận xét: 
- Khi xét mẫu lớn hơn thì độ chính xác đúng hơn và dữ liệu thể hiện rõ ở đoạn [0.125, 0.2] thay vì mẫu dữ liệu tập trung khoảng [~0.07, ~0.22]
